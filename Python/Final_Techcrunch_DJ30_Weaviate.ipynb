```python
from langchain.schema import Document
from langchain.document_loaders import PyPDFLoader
import os
import re

# Define the directory containing PDF files
pdf_directory = "/content/drive/My Drive/P298/data/RAC30/"

# Dow Jones 30 companies for reference
dow_jones_30 = {
    "AAPL": "Apple Inc.",
}

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    loader = PyPDFLoader(pdf_path)
    documents = loader.load()
    # Combine all pages into a single string
    return " ".join(doc.page_content for doc in documents)

# Function to chunk text
def chunk_text(text, chunk_size=1000, overlap=200):
    chunks = []
    for i in range(0, len(text), chunk_size - overlap):
        chunks.append(text[i:i + chunk_size])
    return chunks

# Function to parse file name and get company name and year
def parse_file_name(file_name):
    pattern = re.compile(r'([A-Z]+)_(\w+|[A-Z]+)_(\d{4})\.pdf')
    match = pattern.match(file_name)
    if match:
        ticker = match.group(2)
        year = match.group(3)
        print(ticker)
        company_name = dow_jones_30.get(ticker, "Unknown Company")
        return company_name, year
    return None, None

# Initialize an empty list to hold all documents
documents = []

# Load each PDF document
for pdf_file in os.listdir(pdf_directory):
    if pdf_file.endswith('.pdf'):
        file_path = os.path.join(pdf_directory, pdf_file)
        try:
            # Extract metadata from filename
            company_name, year = parse_file_name(pdf_file)

            # Extract text from PDF
            text = extract_text_from_pdf(file_path)

            # Chunk text
            chunks = chunk_text(text)

            # Append chunks with metadata
            for i, chunk in enumerate(chunks):
                metadata = {
                    "company_name": company_name,
                    "year": year,
                    "chunk_index": i,
                    "file_name": pdf_file
                }
                print(f"Chunking {company_name} {year} - Chunk index {i}")
                documents.append(Document(page_content=chunk, metadata=metadata))

        except Exception as e:
            print(f"Error loading {file_path}: {e}")

# Print the document information
print(f'You have {len(documents)} document(s) in your data')
if documents:
    print(f'There are {len(documents[0].page_content)} characters in your first document')
```

```python
import weaviate
from weaviate.util import get_valid_uuid
from transformers import AutoTokenizer, AutoModel
import torch
import uuid





# Initialize Weaviate client with API key for authentication
#client = weaviate.Client(
#    url="",
#    auth_client_secret=weaviate.AuthApiKey(api_key="")  # Replace with actual API key
#)


client = weaviate.Client(
    url="",
    auth_client_secret=weaviate.AuthApiKey(api_key="")  # Replace with actual API key
)
# Define Weaviate class schema if it doesn't already exist
class_name = "Document"  # Adjust as per your schema

# Check if class already exists, if not, create it
schema = client.schema.get()
if class_name not in [c["class"] for c in schema["classes"]]:
    # Class does not exist; proceed to create it
    client.schema.create_class({
        "class": class_name,
        "properties": [
            {
                "name": "content",
                "dataType": ["text"]
            },
            {
                "name": "metadata",
                "dataType": ["text"]
            },
            {
                "name": "vector",
                "dataType": ["number[]"]  # Adding a property to store embeddings
            }
        ]
    })

# Initialize the embedding model
model_name = "sentence-transformers/all-mpnet-base-v2"  # Choose an embedding model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Function to generate embeddings
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        embeddings = model(**inputs).last_hidden_state.mean(dim=1).squeeze()
    return embeddings.numpy().tolist()

# Function to clean metadata within each document
def clean_metadata(metadata):
    cleaned_meta = {}
    for key, value in metadata.items():
        if value is None:
            cleaned_meta[key] = 'unknown'  # Replace None with 'unknown'
        else:
            cleaned_meta[key] = value
    return cleaned_meta

# Function to batch documents
def batch_documents(documents, batch_size):
    for i in range(0, len(documents), batch_size):
        yield documents[i:i + batch_size]

# Set your batch size
batch_size = 40  # Adjust based on Weaviate's limits

# Process documents in batches and add to Weaviate
for batch in batch_documents(documents, batch_size):
    with client.batch as batcher:
        for doc in batch:
            if doc.metadata:  # Only clean if metadata exists
                doc.metadata = clean_metadata(doc.metadata)
            # Generate vector embedding
            vector = get_embedding(doc.page_content)
            # Generate a unique ID for each document based on its content
            doc_id = str(uuid.uuid5(uuid.NAMESPACE_DNS, doc.page_content))
            # Add document content, metadata, and embedding vector to Weaviate
            batcher.add_data_object(
                data_object={
                    "content": doc.page_content,
                    "metadata": str(doc.metadata),  # Convert metadata to string format
                    "vector": vector
                },
                class_name=class_name,
                uuid=doc_id,
                vector=vector
            )
    # Force batch to process
    batcher.flush()
```

```python
import weaviate
from weaviate.auth import AuthApiKey
from transformers import AutoTokenizer, AutoModel
import torch

# Initialize Weaviate client with API key for authentication
client = weaviate.Client(
    url="",
    auth_client_secret=weaviate.AuthApiKey(api_key="")
)

# Define the query question
query_text = "What is the net profit of Apple in the year of 2023?"

# Load the embedding model
model_name = "sentence-transformers/all-mpnet-base-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Function to generate embeddings
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        embeddings = model(**inputs).last_hidden_state.mean(dim=1).squeeze()
    return embeddings.numpy().tolist()

# Generate embedding for the query
query_embedding = get_embedding(query_text)

# Perform a query on Weaviate with vector search
response = client.query.get(
    class_name="Document",
    properties=["content", "metadata"]
).with_near_vector({"vector": query_embedding}).with_limit(3).do()

# Process and display the response
for result in response["data"]["Get"]["Document"]:
    print("Content:", result["content"])
    metadata = result.get("metadata", {})

    # Check if metadata is a dictionary and print relevant fields
    if isinstance(metadata, dict):
        print("Chunk Index:", metadata.get("chunk_index", "N/A"))
        print("Source File:", metadata.get("source_file", "N/A"))
    else:
        print("Metadata:", metadata)
    print()
```
